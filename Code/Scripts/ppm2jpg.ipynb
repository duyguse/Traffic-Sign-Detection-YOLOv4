**It converts images in .ppm format to .jpg format.**

**If you download TrainIJCNN2013.zip and TestIJCNN2013.zip you can execute following 2 cells**

```
from PIL import Image 
import glob

filepaths = glob.glob("/content/gdrive/My Drive/TrafficSign/Test/ppm/*.ppm")


for path in filepaths:
  im = Image.open(path)
  path = path.replace("ppm","jpg")
  im.save(path)
```
