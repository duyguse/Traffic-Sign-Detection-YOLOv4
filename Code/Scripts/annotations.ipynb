**Annotations converted from GTSDB format to yolo format for 43 class. But it can be integrate to any number of class as you can see next cell.**

```
import os
import json
from os import listdir, getcwd
from os.path import join
import PIL

classes = ["speedlimit20" ,"speedlimit30" ,"speedlimit50" ,"speedlimit60" ,"speedlimit70" ,"speedlimit80" ,"restrictionends80" ,
"speedlimit100" ,"speedlimit120" ,"noovertaking" ,"noovertakingtrucks","priorityatnextintersection","priorityroad" ,"giveway" ,"stop" ,
"notrafficbothways" ,"notrucks" ,"noentry" ,"danger","bendleft","bendright","bend","unevenroad" ,"slipperyroad" ,"roadnarrows" ,"construction",
"trafficsignal","pedestriancrossing" ,"schoolcrossing" ,"cyclescrossing" ,"snow","animals" ,"restrictionends" ,"goright" ,"goleft","gostraight",
"gorightorstraight","goleftorstraight","keepright","keepleft","roundabout","restrictionendsovertaking","restrictionendsovertakingtrucks"]
MAX_WIDTH = 416
MAX_HEIGHT = 416

#box form[x,y,w,h]
def convert(size,box):
  width_proportion = size[0] / MAX_WIDTH
  height_proportion = size[1] / MAX_HEIGHT

  left_x = float(box[0]) / width_proportion
  bottom_y = float(box[1]) / height_proportion
  right_x = float(box[2]) / width_proportion
  top_y = float(box[3]) / height_proportion

  object_width = right_x - left_x
  object_height = top_y - bottom_y
  object_mid_x = (left_x + right_x) / 2.0
  object_mid_y = (bottom_y + top_y) / 2.0

  object_width_rel = object_width / MAX_WIDTH
  object_height_rel = object_height / MAX_HEIGHT
  object_mid_x_rel = object_mid_x / MAX_WIDTH
  object_mid_y_rel = object_mid_y / MAX_HEIGHT

  return (object_mid_x_rel, object_mid_y_rel, object_width_rel, object_height_rel)

def convert_annotation():
  path ='/content/gdrive/My Drive/TrafficSign/Test/'
  with open(path + 'gt.txt','r') as f:
    lines = f.readlines()

  for i, line in enumerate(lines):
    print("{}/{}".format(i, len(lines)))
    line = line.replace("\n", "")
    x = line.split(";")
    x[0] = x[0][:-4]
    outpathtxt = os.path.join(path, 'jpg/%s.txt'%(x[0]))
    outfile = open(outpathtxt, 'w')
    inputImgPath = os.path.join(path, "jpg/"+str(x[0])+".jpg")
    image = PIL.Image.open(inputImgPath)
    width, height = image.size
    bb = convert((width,height),(x[1],x[2],x[3],x[4]))
    info = str(x[5])+" "+" ".join([str(a) for a in bb]) + '\n'
    print("Writing to {} [{}] from {}".format(outpathtxt, info, inputImgPath))
    print("\t\tx was = {}".format(x))
    outfile.write(info)
    outfile.close()
			
if __name__ == '__main__':
    convert_annotation()
    print("done")
```
